```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/AB_NYC_2019.csv.zip')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns=['name','neighbourhood','calculated_host_listings_count'])
df.head()
```

```python
df.info()
```

```python
df['host_name'] = df['host_name'].astype(str)
df['host name'] = df['host_name'].str.replace("nan","  ")
df['host name'] = df['host_name'].str.replace('[^a-zA-Z]','',regex=True)
df['host name'] = df['host_name'].str.replace(r'([a-z])(A-Z)',r'\1\2',regex=True)
df.head()
```

```python
df_priced = df[df['price'] != 0]
df_priced.head()
```

```python
df_year = df[df['minimum_nights'] <= 367]
df_year.head()
```

```python
df_availabel = df[df['availability_365'] !=0]
df_availabel.head()
```

```python
df['last_review'] = df['last_review'].astype(str)
df['last_review'] = df['last_review'].str.replace("nan","  ")
df.head()
```

```python
df['reviews_per_month'] = df['reviews_per_month'].astype(str)
df['reviews_per_month'] = df['reviews_per_month'].str.replace("nan","  ")
df.head()
```
